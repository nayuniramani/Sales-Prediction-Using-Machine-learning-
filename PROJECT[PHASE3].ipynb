```python
#Import mode function:
from scipy.stats import mode

#Determing the mode for each
Store_Capacity_mode = data.pivot_table(values='Store_Capacity', columns='Outlet_Type',aggfunc=lambda x:x.mode())
Store_Capacity_mode
```

```python
def impute_size_mode(cols):
    Size = cols[0]
    Type = cols[1]
    if pd.isnull(Size):
        return Store_Capacity_mode.loc['Store_Capacity'][Store_Capacity_mode.columns == Type][0]
    else:
        return Size

print ('Orignal #missing: %d'%sum(data['Store_Capacity'].isnull()))
data['Store_Capacity'] = data[['Store_Capacity','Outlet_Type']].apply(impute_size_mode,axis=1)
print ('Final #missing: %d'%sum(data['Store_Capacity'].isnull()))
```
